#### Extracting metadata 

```python
import pandas as pd
import os

def load_csv_files(directory):
    """Carga todos los archivos CSV en un diccionario de DataFrames."""
    dataframes = {}
    for file in os.listdir(directory):
        if file.endswith(".csv"):
            file_path = os.path.join(directory, file)
            df_name = file.replace(".csv", "")
            dataframes[df_name] = pd.read_csv(file_path)
            print(f"Cargado: {file} ({df_name})")
    return dataframes

def show_metadata(dataframes):
    """Muestra metadatos de los DataFrames cargados."""
    for name, df in dataframes.items():
        print(f"\n=== {name.upper()} ===")
        print("Columnas:", df.columns.tolist())
        print("Primeras filas:")
        print(df.head())
        print("\nDescripción:")
        print(df.describe(include='all'))

def main():
    directory = "./LeetScrapeData"  # Cambia esto si es necesario
    dataframes = load_csv_files(directory)
    show_metadata(dataframes)

if __name__ == "__main__":
    main()
```

```python
import json
import nbformat as nbf
import pandas as pd
import requests

# Load questions from questions.csv
questions_file = "questions.csv"
df = pd.read_csv(questions_file)

# Function to fetch question details from LeetCode GraphQL API
def get_question_details(title_slug):
    url = "https://leetcode.com/graphql"
    query = {
        "query": """
        query getQuestion($titleSlug: String!) {
            question(titleSlug: $titleSlug) {
                content
            }
        }
        """,
        "variables": {"titleSlug": title_slug}
    }
    response = requests.post(url, json=query)
    if response.status_code == 200:
        data = response.json()
        return data["data"]["question"]["content"] if data["data"]["question"] else "Definition not available"
    return "Failed to fetch data"

# Dictionary of solutions by titleSlug
solutions = {
    "two-sum": """
def twoSum(nums, target):
    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    return []

# Test
twoSum([2,7,11,15], 9)  # [0,1]
""",
    "add-two-numbers": """
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode()
    curr, carry = dummy, 0
    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        carry, out = divmod(val1 + val2 + carry, 10)
        curr.next = ListNode(out)
        curr = curr.next
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    return dummy.next
""",
    "longest-substring-without-repeating-characters": """
def lengthOfLongestSubstring(s: str) -> int:
    char_set = set()
    left, max_length = 0, 0
    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)
    return max_length

# Test
lengthOfLongestSubstring("abcabcbb")  # 3
"""
}

# Create notebook
nb = nbf.v4.new_notebook()
cells = []

# Add introduction cell
cells.append(nbf.v4.new_markdown_cell("""
# LeetCode Questions and Solutions
This notebook contains a list of LeetCode questions and their solutions in Python.
Execute each cell to see the results.
"""))

# Generate cells for each question with its solution
titles = df["QID"].tolist()[:100]  # Take up to 40 questions
for _, row in df.iterrows():
    if len(cells) >= 80:  # Ensure we do not exceed 40 questions (each has two cells)
        break
    qid, slug = row["QID"], row["titleSlug"]
    question_content = get_question_details(slug)
    question_markdown = f"""## Question {qid}: {slug.replace('-', ' ').title()}

LeetCode Exercise: [{slug.replace('-', ' ').title()}](https://leetcode.com/problems/{slug}/)

{question_content}
"""
    
    code_solution = solutions.get(slug, "print('Solution not available')")
    
    cells.append(nbf.v4.new_markdown_cell(question_markdown))
    cells.append(nbf.v4.new_code_cell(code_solution))

# Save the notebook
dst_file = "leetcode_notebook_100.ipynb"
nb["cells"] = cells
with open(dst_file, "w", encoding="utf-8") as f:
    f.write(json.dumps(nb, indent=2))

print(f"Notebook generated: {dst_file}")
```
