```python
import json
import nbformat as nbf
import pandas as pd

# Cargar preguntas desde Questions.csv
questions_file = "Questions.csv"
df = pd.read_csv(questions_file)

# Diccionario de soluciones por títuloSlug
solutions = {
    "two-sum": """
def twoSum(nums, target):
    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    return []

# Prueba
twoSum([2,7,11,15], 9)  # [0,1]
""",
    "add-two-numbers": """
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode()
    curr, carry = dummy, 0
    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        carry, out = divmod(val1 + val2 + carry, 10)
        curr.next = ListNode(out)
        curr = curr.next
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    return dummy.next
""",
    "longest-substring-without-repeating-characters": """
def lengthOfLongestSubstring(s: str) -> int:
    char_set = set()
    left, max_length = 0, 0
    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)
    return max_length

# Prueba
lengthOfLongestSubstring("abcabcbb")  # 3
"""
}

# Crear notebook
nb = nbf.v4.new_notebook()
cells = []

# Agregar una celda de introducción
cells.append(nbf.v4.new_markdown_cell("""
# LeetCode Questions and Solutions
Este notebook contiene una lista de preguntas y sus soluciones en Python.
Ejecuta cada celda para ver los resultados.
"""))

# Generar celdas para cada pregunta con su solución
titles = df[["QID", "titleSlug"]].head(3)  # Tomamos solo 3 preguntas como ejemplo
for _, row in titles.iterrows():
    qid, slug = row["QID"], row["titleSlug"]
    question_markdown = f"""## Pregunta {qid}: {slug.replace('-', ' ').title()}\n\nEjercicio de LeetCode: [{slug.replace('-', ' ').title()}](https://leetcode.com/problems/{slug}/)"""

    
    code_solution = solutions.get(slug, "print('Solución no disponible')")
    
    cells.append(nbf.v4.new_markdown_cell(question_markdown))
    cells.append(nbf.v4.new_code_cell(code_solution))

# Guardar el notebook
dst_file = "leetcode_notebook.ipynb"
nb["cells"] = cells
with open(dst_file, "w", encoding="utf-8") as f:
    f.write(json.dumps(nb, indent=2))

print(f"Notebook generado: {dst_file}")
```

```python
import pandas as pd
import nbformat as nbf

# Load the questions dataset
questions_file = "Questions.csv"
solutions_file = "Solutions.csv"

try:
    df_questions = pd.read_csv(questions_file)
    df_solutions = pd.read_csv(solutions_file)
except FileNotFoundError:
    print("Error: Questions.csv or Solutions.csv not found.")
    exit()

# Ensure required columns exist
if "QID" not in df_questions.columns or "titleSlug" not in df_questions.columns:
    print("Error: Questions.csv is missing required columns.")
    exit()
if "QID" not in df_solutions.columns or "language" not in df_solutions.columns or "solution" not in df_solutions.columns:
    print("Error: Solutions.csv is missing required columns.")
    exit()

# Create a new notebook
nb = nbf.v4.new_notebook()
cells = []

# Add a title cell
cells.append(nbf.v4.new_markdown_cell("""
# LeetCode Questions and Solutions

This notebook contains a list of LeetCode questions with their solutions in multiple programming languages.

Click on each cell to view and run the solutions.
"""))

# Iterate over questions
for _, row in df_questions.iterrows():
    qid, slug = row["QID"], row["titleSlug"]
    title = slug.replace("-", " ").title()
    
    # Markdown cell for the question
    question_markdown = f"""## Question {qid}: {title}

[View on LeetCode](https://leetcode.com/problems/{slug}/)"""
    cells.append(nbf.v4.new_markdown_cell(question_markdown))
    
    # Get all solutions for the current question
    solutions = df_solutions[df_solutions["QID"] == qid]
    
    if not solutions.empty:
        for _, sol in solutions.iterrows():
            language = sol["language"].capitalize()
            code = sol["solution"]
            
            # Add a markdown header for the language
            cells.append(nbf.v4.new_markdown_cell(f"### Solution in {language}"))
            
            # Add a code cell with the solution
            cells.append(nbf.v4.new_code_cell(code))
    else:
        cells.append(nbf.v4.new_markdown_cell("**No solutions available for this question.**"))

# Add all cells to the notebook
nb["cells"] = cells

# Save the notebook
notebook_filename = "LeetCode_Solutions.ipynb"
with open(notebook_filename, "w", encoding="utf-8") as f:
    nbf.write(nb, f)

print(f"Notebook '{notebook_filename}' successfully generated!")
```

```python
from leetscrape import GetQuestionsList, GetQuestion
import pandas as pd

# Obtener la lista de preguntas
ls = GetQuestionsList()
ls.scrape()

# Filtrar solo preguntas gratuitas
questions_df = ls.questions[ls.questions["paidOnly"] == False]

# Función para obtener el enunciado de cada pregunta
def get_problem_statement(title_slug):
    try:
        q = GetQuestion(titleSlug=title_slug).scrape()
        return q.Body
    except Exception as e:
        return f"Error: {e}"

# Aplicar la función a todas las preguntas gratuitas
questions_df["ProblemStatement"] = questions_df["titleSlug"].apply(get_problem_statement)

# Guardar en CSV
questions_df.to_csv("leetcode_questions_with_statements.csv", index=False)
```
